## Resize

```python
def resize(image, target_size):

    original_height, original_width, _ = image.shape
    target_width, target_height = target_size
    resized = np.zeros((target_height, target_width, 3), dtype=np.uint8)
    
    for i in range(target_height):
        for j in range(target_width):
            original_x = int(j * original_width / target_width)
            original_y = int(i * original_height / target_height)
            resized[i, j] = image[original_y, original_x]
    
    return resized
```

Cell code di atas merupakan implementasi manual dari fungsi resize citra menggunakan metode nearest neighbor interpolation, tanpa memanfaatkan fungsi resize bawaan dari library cv2. Fungsi ini bekerja dengan cara membuat array kosong baru ("resized") berukuran sesuai target_size yang diinginkan, kemudian mengisi setiap piksel pada array tersebut melalui dua buah perulangan for yang menyusuri tinggi (i) dan lebar (j) citra target. Untuk setiap posisi piksel pada citra hasil, dilakukan perhitungan koordinat asal (original_x dan original_y) pada citra original dengan cara mengalikan posisi piksel target dengan rasio antara ukuran original dan ukuran target, sehingga nilai piksel dari citra asli pada koordinat tersebut dapat diambil dan dipetakan ke posisi yang bersesuaian pada citra hasil resize.

## Data Loading

Cell code di atas merupakan proses pembacaan dan pemuatan seluruh citra yang terdapat pada folder "Assets" untuk kemudian diolah lebih lanjut. Pertama, dilakukan perulangan terhadap setiap sub_folder yang ada di dalam folder Assets, di mana setiap sub_folder ini diasumsikan merepresentasikan satu label atau kelas tertentu. Selanjutnya, setiap file citra yang berada di dalam masing-masing sub_folder dibaca menggunakan cv.imread, kemudian diubah ukurannya menjadi 480x640 piksel menggunakan fungsi resize yang telah didefinisikan sebelumnya, dan diubah pula menjadi citra grayscale menggunakan cv.cvtColor agar proses ekstraksi fitur pada tahap berikutnya menjadi lebih sederhana karena hanya berfokus pada satu channel intensitas piksel saja.

## Preprocesing

### Mendefinisikan Fungsi

```python
def prepro1(image):
    img_min = np.min(image)
    img_max = np.max(image)
    if img_max == img_min: 
        norm = image.astype(np.uint8)
    else :
        norm = ((image - img_min) / (img_max - img_min) * 255).astype(np.uint8)

    h, w = norm.shape
    pad = 3 // 2
    padded = np.pad(norm, pad, mode='constant', constant_values=0)
    hasil = np.zeros((h, w), dtype=np.uint8)
    
    for i in range(h):
        for j in range(w):
            window = padded[i:i+3, j:j+3]
            hasil[i, j] = np.median(window)
    return hasil
```

Cell code di atas merupakan fungsi prepro1 yang dibuat untuk melakukan preprocessing citra melalui dua tahapan, yaitu normalisasi intensitas piksel dan median filtering. Pada tahap pertama, dilakukan normalisasi terhadap citra dengan cara mengambil nilai minimum (img_min) dan maksimum (img_max) dari citra, kemudian memetakan ulang seluruh nilai piksel ke dalam rentang 0 hingga 255 menggunakan rumus normalisasi, dengan pengecualian apabila nilai minimum dan maksimum sama maka citra langsung dikonversi ke tipe data uint8 tanpa proses normalisasi untuk menghindari pembagian dengan nol. Selanjutnya, pada tahap median filtering, citra yang telah dinormalisasi tersebut diberikan padding sebesar 1 piksel di setiap sisinya menggunakan np.pad agar piksel pada bagian tepi citra tetap dapat diproses, kemudian dilakukan perulangan terhadap setiap piksel untuk mengambil window berukuran 3x3 di sekitarnya dan menghitung nilai mediannya menggunakan np.median, sehingga nilai piksel pada citra hasil ("hasil") merupakan nilai tengah dari kumpulan piksel tetangganya yang berguna untuk mengurangi noise pada citra sambil tetap mempertahankan ketajaman tepi objek.

### Menjalankan Fungsi

```python
def jalankan_proses(arraydata, fungsi_prepro):
    data_hasil = []
    for img in arraydata:
        img_hasil = fungsi_prepro(img)
        data_hasil.append(img_hasil)
    return np.array(data_hasil)
```

Cell code di atas merupakan fungsi pembantu (helper function) bernama jalankan_proses yang dibuat untuk menerapkan suatu fungsi preprocessing tertentu, seperti prepro1, ke seluruh citra yang ada di dalam array data secara sekaligus. Fungsi ini menerima dua parameter, yaitu arraydata yang merupakan kumpulan citra yang ingin diproses, dan fungsi_prepro yang merupakan fungsi preprocessing yang ingin diterapkan, sehingga fungsi ini bersifat fleksibel dan dapat digunakan kembali untuk berbagai fungsi preprocessing lainnya tanpa perlu menulis ulang perulangan yang sama. Di dalam fungsi ini, dilakukan perulangan terhadap setiap citra pada arraydata, di mana setiap citra tersebut diproses menggunakan fungsi_prepro yang diberikan, kemudian hasilnya dimasukkan ke dalam list data_hasil, yang pada akhirnya dikonversi menjadi array numpy dan dikembalikan sebagai output dari fungsi.

### Menampilkan hasil proses preprosessing 1

```python
data_prepro1 = jalankan_proses(data, prepro1)
displayImg(10, 20, data_prepro1)
```

Cell code di atas merupakan implementasi pemanggilan fungsi jalankan_proses untuk menerapkan preprocessing prepro1 ke seluruh citra pada variabel "data", yang hasilnya disimpan ke dalam variabel baru bernama data_prepro1, kemudian ditampilkan menggunakan fungsi displayImg yang telah dibuat sebelumnya. Apabila dibandingkan dengan citra sebelum diproses, citra hasil prepro1 ini secara visual akan terlihat sedikit lebih halus (smooth) dan noise atau bintik-bintik kecil yang sebelumnya tampak pada citra original akan tereduksi akibat proses median filtering, sementara kontras antar objek pada citra juga akan tampak lebih jelas dan merata dibandingkan citra aslinya karena telah melalui proses normalisasi intensitas piksel ke rentang 0-255, sehingga perbedaan tingkat kecerahan antar citra yang sebelumnya bisa jadi tidak konsisten kini menjadi lebih seragam satu sama lain.
